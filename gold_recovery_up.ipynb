**Технологический процесс**

* Rougher feed — исходное сырье

```python
#**Функция print_basic_info, для вывода информации о массиве, и его переменных.**

#4 входных параметра

#* base - название базы данных
#* info - 1: вывод информации о массиве, другое: не вывод
#* head - n: вывод примера базы (вывод n - строк), n < 1: не вывод
#* describe - 1: вывод описания переменных массива, другое: не вывод        

def print_basic_info(base, info, head, describe):
    if info == 1:
        print("\n", bold(), colored('info','green'), bold_end(), "\n")
        print( base.info())  
    if head >= 1:
        print("\n", bold(),colored('head','green'),bold_end())
        display(base.head(head))
    if describe == 1:
        print("\n", bold(),colored('describe','green'),bold_end(),"\n")
        for i in base.columns:
            print("\n", bold(), colored(i,'blue'),bold_end(),"\n", base[i].describe())
                   
#Функция ___ft_graph_info___, для вывода названия и лейбла переменной, частотной нормированной таблицы и описания переменной.

#*5 входных параметров*

#+ *base* - название базы данных
#+ *name* - имя переменной в базе
#+ *describe* - 1: вывод описания переменной и границ значений без выбросо, 0: не вывод
#+ *boxplot* - 1: вывод диаграммы boxplot в, 0: не вывод
#+ *hist* - 1: вывод гистограммы, без экстримальных значений, 0: не вывод

def ft_graph_info(base, name, describe, boxplot, hist ):
    IQR = (base[name].quantile(.75) - base[name].quantile(.25)) * 1.5
    min_range = base[name].quantile(.25) - IQR
    max_range = base[name].quantile(.75) + IQR
    if min_range < 0:
        min_range = 0
    print (name,"\n")
    if describe ==1:
        print(base[name].describe())
    if boxplot ==1:
        print ("\n Границы значений без экстремальных выбросов: ", 
               int(round(min_range,0)),
               "-", int(round(max_range,0)))
        base.boxplot(name)
        plt.show()
    if hist ==1:
        base.hist(name, bins = 50 , range = (min_range , max_range))
        
# Графиики для наглядности

#*2 входных параметра*

#+ *base* - название базы данных
#+ *name* - название графика

def param_bars(base_name, name):
    sns.set(style="whitegrid")
    plt.figure(figsize = (10,3)) 
    df = base_name
    sns.barplot(data=df, palette="tab20", linewidth=2.5)
    plt.title( str(name), fontsize=15)
    plt.ylabel("%%")
    plt.xlabel("Параметры")
    plt.ylim((0, 10)) 
    c = 0
    for i in df.columns:
        plt.text( c - 0.05 , df[i].mean() + 0.1, str(round(df[i].mean(),1)) , fontsize=15)
        c = c + 1
    plt.show()
```

**Проверьте, что эффективность обогащения рассчитана правильно. Вычислите её на обучающей выборке для признака rougher.output.recovery.**

```python
#Выделим нужные столбцы, уберем пропуски
df_rougher_check = df_train[df_train.columns[30:65]].dropna(subset=['rougher.output.recovery'])

df_rougher_check.info()
```

```python
#Посмотрим распределение
df_rougher_check['rougher.output.recovery'].hist(bins = 50, alpha=0.7, density=True)
plt.xlabel('% обогащения')
plt.ylabel('Плотность распределения')
plt.title('Эффективность обогащения "rougher.output.recovery"', fontsize=15) 
plt.show()
```

```python
#Напишем функцию рековери
C = df_rougher_check['rougher.output.concentrate_au'] 
F = df_rougher_check['rougher.input.feed_au']
T = df_rougher_check['rougher.output.tail_au']

def recovery_check(C, F, T):  return C*(F-T) / (F*(C-T)) * 100

#Посчитаем
df_rougher_check['rougher.output.recovery_check'] = recovery_check(C,F,T)
```

```python
# Посмотрим на медианы 
print(df_rougher_check['rougher.output.recovery'].median(), "Vs"
      ,df_rougher_check['rougher.output.recovery_check'].median())

#А смап?
def smape(A, F):
    return 100/len(A) * np.sum(2 * np.abs(F - A) / (np.abs(A) + np.abs(F)))

print(smape(df_rougher_check['rougher.output.recovery_check'], df_rougher_check['rougher.output.recovery']))
```

```python
def fin_smape(df_target, df_predict):
    return (0.25 * smape(df_target['rougher.output.recovery'], df_predict['rougher.output.recovery']) 
    +  (0.75 * smape(df_target['final.output.recovery'], df_predict['final.output.recovery'])))

def fin_s(smape_r, smape_f):
    return ((0.25 * smape_r)  + ( 0.75 * smape_f))
```

```python
# инициализация KFold с 5тью фолдами
cv = KFold(n_splits=5, random_state=1234, shuffle=True)

ans_up = pd.DataFrame()

#for clf in ((LinearRegression()),(LinearRegression())):
for clf in ((LinearRegression()),
            (AdaBoostRegressor(n_estimators = 15, loss = 'exponential', learning_rate = 0.2)), 
            (BayesianRidge(n_iter = 1)),
            (GradientBoostingRegressor (n_estimators = 181, loss = 'lad', learning_rate = 0.5))):
    clf = MultiOutputRegressor(clf)
    n = 0
    base = pd.DataFrame()
    for train_index, val_index in cv.split(y_train_val):
        X_train, X_val = X_train_val.loc[train_index], X_train_val.loc[val_index]
        y_train, y_val = y_train_val.loc[train_index], y_train_val.loc[val_index]
        #print(X_train.shape, X_val.shape,y_train.shape, y_val.shape)
        
        clf.fit(X_train, y_train)
        tmp = pd.DataFrame(data = clf.predict(X_val), columns = ['rougher.output.recovery', 'final.output.recovery'])
        
        #print(tmp.shape, y_val['final.output.recovery'].shape,tmp['final.output.recovery'].shape)
        base.loc[n,'smape_r'] = smape(y_val['rougher.output.recovery'], tmp['rougher.output.recovery'])
        base.loc[n,'smape_f'] = smape(y_val['final.output.recovery'], tmp['final.output.recovery'])
        base.loc[n,'fin_smape'] = fin_smape(y_val, tmp)
        base.loc[n,'fin_smape'] = fin_smape(y_val, tmp)
        
        #print(y_val['final.output.recovery'].shape,tmp['final.output.recovery'].shape)
        base.loc[n,'mae_f'] = mean_absolute_error(y_val['final.output.recovery'], tmp['final.output.recovery'])**0.5
        base.loc[n,'mse_f'] = mean_squared_error(y_val['final.output.recovery'], tmp['final.output.recovery'])**0.5
        
        base.loc[n,'mae_r'] = mean_absolute_error(y_val['rougher.output.recovery'], tmp['rougher.output.recovery'])**0.5
        base.loc[n,'mse_r'] = mean_squared_error(y_val['rougher.output.recovery'], tmp['rougher.output.recovery'])**0.5
        
        n = n +1
    ans_up.loc[str(clf).split('(')[1].split('=')[1],'mae_r'] = base['mae_r'].mean()
    #ans_up.loc[str(clf).split('(')[1].split('=')[1],'mse_r'] = base['mse_r'].mean()
    ans_up.loc[str(clf).split('(')[1].split('=')[1],'mae_f'] = base['mae_f'].mean()
    #ans_up.loc[str(clf).split('(')[1].split('=')[1],'mse_f'] = base['mse_f'].mean()
    ans_up.loc[str(clf).split('(')[1].split('=')[1],'smape_r'] = base['smape_r'].mean()
    ans_up.loc[str(clf).split('(')[1].split('=')[1],'smape_f'] = base['smape_f'].mean()
    ans_up.loc[str(clf).split('(')[1].split('=')[1],'fin_smape'] = base['fin_smape'].mean()
```
